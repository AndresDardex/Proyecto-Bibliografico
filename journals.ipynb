```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup

from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt

import string
import nltk
from nltk.corpus import stopwords
```

```python
data = pd.read_csv("Search.csv")
```

```python
Articles = data[['Source title','Document Type']]
```

```python
Articles = Articles.groupby('Document Type').get_group('Article')
```

```python
source = pd.DataFrame(Articles['Source title'].value_counts())
```

```python
source = source.reset_index()
```

```python
source = source.rename(columns={'index':'Source title', 'Source title':'Count'})
```

```python
source.head(10)
```

```python
counts = source.head(8)
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = 'grey'
```

```python
fig, ax = plt.subplots()
ax.barh(counts['Source title'], counts['Count'], color = 'black')

ax.set_title('Número de artículos por Journal')
ax.set_xlabel('Número de artículos')
ax.set_ylabel('Journal')

plt.show()
```
